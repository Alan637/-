```python
import pandas as pd
import numpy as np
df=pd.read_csv(r'D:/2016-FCC-New-Coders-Survey-Data.csv',usecols=['CityPopulation','EmploymentStatus','Gender','HasDebt','JobPref','JobWherePref','MaritalStatus','Income','SchoolDegree'])
```

```python
df
```

```python
df=df.dropna()
df=df[(df.Gender=='male')|(df.Gender=='female')]
```

```python
df
```

```python
import scipy.stats as stats
stats.shapiro(df.Income)
stats.shapiro(df.Income.apply(lambda x: np.log(x)))
stats.shapiro(df.Income[0:100].apply(lambda x: np.log(x)))
df2=df.iloc[0:100,:].copy()
df2.Income=df2.Income.apply(lambda x: np.log(x))
from statsmodels.formula.api import ols
model = ols('Income ~ SchoolDegree',df).fit()
model.summary()
model = ols('Income ~ SchoolDegree',df2).fit()
model.summary()
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd, MultiComparison
mc=MultiComparison(df2.Income,df2.SchoolDegree)
print(mc.tukeyhsd())
```
